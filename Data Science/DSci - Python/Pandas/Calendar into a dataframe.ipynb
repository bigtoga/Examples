```python
start_date = "2020/01/01"
end_date = "2020/04/20";

customCalendar = pd.DataFrame({"Date": pd.date_range(start_date, end_date)})
customCalendar["day_of_week"] = customCalendar.Date.dt.weekday_name
customCalendar["day_of_month"] = customCalendar.Date.dt.day
customCalendar["week"] = customCalendar.Date.dt.weekofyear
customCalendar["month"] = customCalendar.Date.dt.month
customCalendar["month_name"] = customCalendar.Date.dt.month_name() # requires the ()
customCalendar["Quarter"] = customCalendar.Date.dt.quarter
customCalendar["Year"] = customCalendar.Date.dt.year
customCalendar["Year_half"] = (customCalendar.Quarter + 1) // 2
customCalendar.index.name = 'id'
customCalendar
```

```python
# Create a datetime variable for today
base = datetime.datetime.today()
# Create a list variable that creates 365 days of rows of datetime values
date_list = [base - datetime.timedelta(days=x) for x in range(0, 365)]

# Create an empty dataframe
df = pd.DataFrame()
# Create a column from the datetime variable
df['datetime'] = date_list

df.head()
```
