```python
import pandas as pd, numpy as np
```

```python
df = pd.read_csv('https://data.princeton.edu/wws509/datasets/salary.dat', delim_whitespace=True)
df.columns = ['sex', 'rank', 'year_at_rank', 'degree', 'year_since_earning_degree', 'salary']
df.head()
```

```python
# hot encoded dummies
# create a new dataset and splits all the possible values of 
# your input data into new columns containing numerical data
# Note that get_dummies() drops the original column and replaces it with 
# "1 new column for each unique value in the column"
categorical_columns = ['sex', 'degree']
dfMultiple = pd.get_dummies(df, columns = categorical_columns) # , drop_first = True)
dfSex = pd.get_dummies(df, columns = ['sex'])
dfDegree = pd.get_dummies(df, columns = ['degree'])
dfSex.head()
```
